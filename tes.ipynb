```python
#panggil library
import numpy as np
import matplotlib.pyplot as tp
#buat data matriks
jd=100
x=np.array([i*0.1 + np.random.randn() for i in range (jd)])
y=np.array([i*0.1 for i in range (jd)])
#buat fungsi linier
def  fi(m,x):
    y=m*x
    return y
#buat perhitungan awal linier
x_prediksi=np.array([0.10])
m_prediksi=1
y_prediksi=fi(x_prediksi, m_prediksi)
#tampilkan pola
tp.scatter(x,y)
tp.plot(x_prediksi, y_prediksi, "red")
tp.axis([0,10,0,10])
tp.show
```
